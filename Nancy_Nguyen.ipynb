```python
def mean_minus_2std(data):
    n = len(data)
    # compute mean
    mean = sum(data) / n
    # compute variance
    variance = 0
    for x in data:
        variance += (x - mean) ** 2
    variance = variance / n
    # compute standard deviation
    std = variance ** 0.5
    return mean - 2 * std
```
